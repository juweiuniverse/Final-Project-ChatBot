### Data Preprocess

```python
from langchain_community.document_loaders import CSVLoader

def load_and_process_data():
    """Memuat data lagu dari CSV ke format dokumen LangChain."""
    print("\n--- Loading and Processing Song Data ---")
    song_docs = []

    try:
        song_loader = CSVLoader(file_path='/content/1200_song_mapped.csv', encoding='utf-8')
        song_docs = song_loader.load()
    except FileNotFoundError as e:
        print(f"Error: {e}. Pastikan file CSV sudah ada di path yang benar.")
        return None
    except RuntimeError as e:
        print(f"Runtime error saat memuat CSV: {e}")
        return None

    print("Data lagu berhasil dimuat.")
    return song_docs
```

```python
# Prepare data and retrievers
song_docs = load_and_process_data()
if not (song_docs):
    exit("Data loading failed. Exiting.")
```

## Vectorized Data Storing

```python
import os
import time # Import time for delays
from langchain.text_splitter import RecursiveCharacterTextSplitter
from langchain_community.vectorstores import FAISS
from tqdm.auto import tqdm # Import tqdm for progress bar
from google.colab import userdata
import litellm
```

```python
print("\n--- Ingesting Data ---")

# Pisahkan dokumen lagu jadi potongan kecil
text_splitter = RecursiveCharacterTextSplitter(chunk_size=1000, chunk_overlap=200)
song_splits = text_splitter.split_documents(song_docs)

print(f"  - Created {len(song_splits)} chunks for Song Dataset.")
```

```python
from langchain_huggingface import HuggingFaceEmbeddings

embeddings = HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2")
```

```python
def vectorize_songs(song_splits, embeddings, batch_size=1000):
    """Vectorizes song documents and stores them in a FAISS vector store."""
    print("\n  - Vectorizing song documents...")
    max_retries = 5
    retry_delay = 10
    song_vector_store = None

    for attempt in range(max_retries):
        try:
            print(f"    Vectorizing songs in batches of {batch_size}...")
            for i in tqdm(range(0, len(song_splits), batch_size)):
                batch = song_splits[i:i + batch_size]
                if song_vector_store is None:
                    song_vector_store = FAISS.from_documents(documents=batch, embedding=embeddings)
                else:
                    song_vector_store.add_documents(batch, embedding=embeddings)
                # Delay kecil untuk aman
                time.sleep(0.5)

            print("  - Song vector store created.")
            return song_vector_store.as_retriever(k=5)

        except Exception as e:
            if "RateLimitError" in str(e) and attempt < max_retries - 1:
                print(f"Rate limit exceeded. Retrying in {retry_delay} seconds... (Attempt {attempt + 1}/{max_retries})")
                time.sleep(retry_delay)
            else:
                print(f"An error occurred during song vector store creation: {e}")
                return None
```

```python
song_retriever = vectorize_songs(song_splits, embeddings)
print("Retriever ready for queries!")
```

## Define Tools

```python
from pydantic import BaseModel
from crewai.tools import BaseTool

# 1. Schema untuk input tool
class SongSearchToolSchema(BaseModel):
    query: dict

# 2. Tool
class SongSearchTool(BaseTool):
    name: str = "Song Search Tool"
    description: str = "Retrieves song info from FAISS vectorstore for a query."
    args_schema = SongSearchToolSchema
    song_retriever: object

    def _run(self, query: dict) -> str:
        # gabungkan dict menjadi format teks
        q = " ".join(f"{k}:{v}" for k, v in query.items())

        # retriever modern LangChain pakai .invoke()
        results = self.song_retriever.invoke(q)

        return "\n".join(doc.page_content for doc in results)
```

```python
song_tool = SongSearchTool(song_retriever=song_retriever)
```

## Define Agents

```python
from crewai import Agent

def define_song_agents(llm, song_tool, web_tool=None):
    """Defines multi-agent team for song recommendation."""

    researcher = Agent(
        role="Song Researcher",
        goal="Find songs relevant to the user's query based on genre, mood, artist, and year.",
        backstory="A music expert skilled in retrieving songs from large datasets.",
        tools=[song_tool],
        llm=llm,
        verbose=True
    )

    validator = None
    if web_tool:
        validator = Agent(
            role="Music Validator",
            goal="Check for recent reviews or news related to songs/artists.",
            backstory="A meticulous music reviewer who cross-checks songs with web data.",
            tools=[web_tool],
            llm=llm,
            verbose=True
        )

    synthesizer = Agent(
        role="Recommendation Synthesizer",
        goal="Create a clear and coherent list of song recommendations integrating retrieved results.",
        backstory="A music curator skilled at producing understandable and engaging song recommendations.",
        llm=llm,
        verbose=True
    )

    return researcher, validator, synthesizer
```

## Define Task

```python
from crewai import Task

def define_song_tasks(researcher, validator=None, synthesizer=None):
    """Defines the sequence of tasks for Spotify song recommendation."""

    research_task = Task(
        description='Search for songs relevant to the user query, considering genre, mood, artist, and year.',
        expected_output='A compiled list of songs matching the query criteria.',
        agent=researcher
    )

    validation_task = None
    if validator:
        validation_task = Task(
            description='Search the web for reviews, charts, and news related to the songs/artists.',
            expected_output='A summary of recent information and opinions from reliable music sources.',
            agent=validator
        )

    synthesis_task = Task(
        description=(
            'Analyze the retrieved songs and any additional validation info. '
            'Synthesize them into a clear and coherent list of recommended songs for the user query.'
        ),
        expected_output='A final curated list of song recommendations ready to present to the user.',
        agent=synthesizer,
        context=[research_task] + ([validation_task] if validation_task else [])
    )

    return [research_task] + ([validation_task] if validation_task else []) + [synthesis_task]
```

## Assembling the Crew

```python
from crewai import LLM
import os

hf_api_key = os.environ.get("HUGGINGFACE_API_KEY")

if not hf_api_key:
    raise ValueError("HUGGINGFACE_API_KEY is not set. Please set it in Colab secrets.")

llm = LLM(
    model="huggingface/meta-llama/Llama-3.1-8B-Instruct",
    api_key= hf_api_key
)

print("LLM initialized successfully.")
```

```python
# 1 Buat agents dari function kamu
researcher, validator, synthesizer = define_song_agents(llm, song_tool=song_tool, web_tool=None)

# 2 Buat tasks khusus Spotify
tasks = define_song_tasks(researcher, validator, synthesizer)
```

```python
from crewai import Crew, Process

# Assemble the Crew
song_recommendation_crew = Crew(
    agents=[researcher, synthesizer],  # pastikan ini Agent object
    tasks=tasks,                        # list of Task
    process=Process.sequential,          # jalankan task secara berurutan
    verbose=True
)
```
